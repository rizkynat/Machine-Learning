```python
import sklearn
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
x = iris.data#atribut
y = iris.target#label
```

```python
from sklearn import metrics#for check error accuracy model
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(model, x,y,cv=5)#cv is fold-n
score
```

if result from cross_val_score is 0.85 - 0.99, so models is good
